```python
from GoogleNews import GoogleNews
from newspaper import Article
import pandas as pd
import numpy as np
from datetime import date
```

```python
# Automated dates starting today and going back 3 years
start_date ='1/1/{}'.format(date.today().year-3)
end_date = '{}/{}/{}'.format(date.today().month, date.today().day,date.today().year) # Do not change this is today's date
start_date, end_date
```

```python
#Initialize GoogleNews() object
googlenews = GoogleNews(start_date, end_date)
```

```python
# Set search term(s)
search_term = 'Seattle SPD'
googlenews.search(search_term)
```

```python
#See Results
result = googlenews.result(sort=True)

df_1=pd.DataFrame(result).drop(['img', 'link'], axis=1)
df_1.dtypes
```

### Each google news such produces the number of articles on a page (up to 10).  In order to get more information, we need to:
1. Iterate through the number of pages wanted, for examle 50 for 500 results
2. Turn each resulting result into a dataframe 
3. Add dataframe to a list
4. Concat all all dataframes in list into one large dataframe

```python
googlenews.getpage(50)
test=googlenews.result()
pd.DataFrame(test).drop('img', axis=1)
```

```python
#Create list to hold search results as dataframes
results_df_list = [df_1] #seed with first set of results above

#We start the iteration at 2 because we already have the first page.  Let's start with 500 total results
for i in range(2,50): #inclusive
    googlenews.getpage(i)
    next_result= googlenews.result()
    #if next_result.dtype != nonetype:
    results_df_list.append(pd.DataFrame(next_result).drop(['img','link'], axis=1)) #Drop the img column as it's unnecessary

#Create a single dataframe by concatenating all dataframes in the list
full_df = pd.concat(results_df_list)
```
