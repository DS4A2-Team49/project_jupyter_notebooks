```python
#import libraries
from GoogleNews import GoogleNews
from newspaper import Article
import pandas as pd
import numpy as np
from datetime import date

#Display full dataFrames
pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
def get_news():
    # Automated dates starting today and going back 3 years
    start_date ='1/1/{}'.format(date.today().year-3)
    end_date = '{}/{}/{}'.format(date.today().month, date.today().day,date.today().year) # Do not change this is today's date

    # Initialize GoogleNews() object
    googlenews = GoogleNews(start_date, end_date)
    
    # Set search term(s)
    search_term = 'Seattle SPD'
    googlenews.search(search_term)
    
    return googlenews
```

```python
def convert_news(start=2, finish=50):
    results_df_list = []
    googlenews = get_news()
    
    for i in range(start,finish): #inclusive
        googlenews.getpage(i)
        news_df= pd.DataFrame(googlenews.result())[['title','media','desc']]
        results_df_list.append(news_df)
        
    return results_df_list
```

```python
def combine_save_df():
    all_news = convert_news().drop_duplicates()
    all_news.to_csv("news_file_spd.csv")
    print('News file has been saved to \"data/news_file_spd.csv\"')
    return all_news
```
