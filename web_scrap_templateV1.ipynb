### Save the page as an html file and use the following code

```python
#Enter the file name
with open('nwasianweekly_05282021.html', 'r') as page_file:
    page_html = page_file.read()
```

### Run the following Code

```python
#html parsing
    soup= BeautifulSoup(page_html, 'lxml')

#Verify that you got the data
print(soup.prettify())
```

### This is the point where you will need to look at the developer tools portion of the website to determine which containers hold the news headlines, their date/timestamp and the news summary.

```python
#Using the developers tool I tested the container that holds the news items
soup.article.text
```

In my example, 'article' is the container for all headline information


```python

headline_containers=soup.find_all('article')

#Check of how many headlines you have captured.  Each headline is in a container
len(headline_containers)
```

### All headlines have the 'h2' tag

```python
#Create a List of Headlines in the variable 'lead'
lead =[]
headlines =soup.find_all('h2')
for headline in headlines:
    if len(headline)>0: #Eliminates empty strings
        lead.append(headline.text)
lead
```

### All new summaries are in a 'div' tag and class ='entry-content'

```python
#Create a List of news summaries in the variable 'news_summaries'
news_summaries =[]
summaries =soup.find_all('div', class_='entry-content')
for summary in summaries:
    news_summaries.append(summary.text)
news_summaries
```

```python
#Create a List of website source in the variable 'sources'
sources =[]
websites =soup.find_all('span', class_='entry-author-name')
for website in websites:
    sources.append(website.text)
sources
```

```python
#Create a List of news summaries in the variable 'news_summaries'
dates =[]
article_dates =soup.find_all('time')
for date in article_dates:
    dates.append(date.text)
dates
```

## Do not change the following code

```python
#Check the length of your lists because it should all be the same
column_lists= [dates, lead, news_summaries, sources]

for col_list in column_lists:
    print(len(col_list))
```

```python
page_columns= {'date':dates,'headline': lead, 'news_summary':news_summaries,'source': sources}
df = pd.DataFrame.from_dict(page_columns)
df
```

```python
#Save DataFrame to csv file
df.to_csv(my_file, index=False)
```
